```python
import pandas as pd
df = pd.read_csv("English.csv")
df.head()
```

```python
df.shape
```

```python
data1 = []
```

```python
for i in range(175621):
    data1.append(df.loc[:,"English words/sentences"][i])
data1
```

```python
len(data1)
```

```python
#French Translation
#LENGH IS TOO LONG 175621, SO IT IS GIVING FOLLOING ERROR. SO I LIMIT THE SIZE TO 100
#RuntimeError: generator raised StopIteration
from translate import Translator
FRENCH_TRANSLATOR= Translator(to_lang="French")
FRENCH_TRANSLATION = []
for i in range(100):
    FRENCH_TRANSLATION.append(FRENCH_TRANSLATOR.translate(data1[i]))
print(FRENCH_TRANSLATION)
```
